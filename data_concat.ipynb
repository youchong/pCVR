```python
import pandas as pd
import numpy as np
import random as rnd

import gc
```

```python
data2_train=pd.read_csv('Features/raw/data2_train.csv')
counts_train=pd.read_csv('Features/counts/data2_counts_train.csv')
timediff_train=pd.read_csv('Features/timediff/data2_timediff_train.csv')
convrate_train=pd.read_csv('Features/convrate/data2_convrate_train.csv')
```

```python
#data2_train.drop(['clickTime','conversionTime'],axis=1,inplace=True)
label=data2_train['label']
data2_train.drop(labels=['label'],axis=1,inplace = True)
data2_train.insert(0,'label',label)
train=pd.concat([data2_train,counts_train,timediff_train,convrate_train],axis=1)
```

```python
train.head()
```

```python
train.to_csv('Features/finaldata/train_raw.csv',index=False)
```

```python
del data2_train
del counts_train
del timediff_train
del convrate_train
del train
del label
gc.collect()
```
