```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd
```

```python
data1=pd.read_csv('Features/raw/data1.csv')
data2=pd.read_csv('Features/raw/data2.csv')
```

```python
data1.head()
```

```python
data1.columns.values
```

```python
for value in ['connectionType', 'creativeID',
       'positionID', 'telecomsOperator', 'userID', 'adID',
       'camgaignID', 'advertiserID', 'appID', 'appPlatform', 'appCategory',
       'age', 'gender', 'education', 'marriageStatus', 'haveBaby',
       'hometown', 'residence', 'sitesetID', 'positionType']:
    data2[value+'Counts']=data2[value].map(data1[value].value_counts())
    data2[value+'Counts'].fillna(0,inplace=True)
    data2[value+'Counts']=data2[value+'Counts'].astype(int)
    data2.drop([value],axis=1,inplace=True)
```

```python
data2.drop(['clickTime','conversionTime','label'],axis=1,inplace=True)
```
