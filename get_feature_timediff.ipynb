```python
# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd
```

```python
data=pd.read_csv('Features/raw/allIDdata.csv')
```

```python
# for value in ['creativeID', 'positionID', 'adID',
#        'camgaignID', 'advertiserID', 'appID']:
value='advertiserID'
data['first']=0
data['last']=0

filename='us'+value[0:2]+'v'

data['temp']=data['userID']*100000+data[value]
data.loc[data.duplicated(['temp'],'last'),'first']=1
data.loc[data.duplicated(['temp'],'first'),'last']=1

firsttime=data[(data['first']==1)&(data['last']==0)][['temp','clickTime']]
lasttime=data[(data['first']==0)&(data['last']==1)][['temp','clickTime']]

firsttime.rename(columns={'clickTime':'ftime'},inplace=True)
lasttime.rename(columns={'clickTime':'ltime'},inplace=True)

data=pd.merge(data,firsttime,how='left')
data=pd.merge(data,lasttime,how='left')

data['ftime'].fillna(0,inplace=True)
data['ftime']=data['ftime'].astype(int)

data['ltime'].fillna(0,inplace=True)
data['ltime']=data['ltime'].astype(int)

data['us'+value[0:2]+'vfirst']=data['clickTime']-data['ftime']
data['us'+value[0:2]+'vlast']=data['ltime']-data['clickTime']

data.loc[(data['first']==0)&(data['last']==0),'us'+value[0:2]+'vfirst']=-1
data.loc[(data['first']==0)&(data['last']==0),'us'+value[0:2]+'vlast']=-1
```
