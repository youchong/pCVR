```python
data2['installedusCounts']=0
data2['installedapCounts']=0
```

```python
for i in ['1','2','3']:
    name='Features/installedapps/installed'+i+'.csv'
    user_installedapps=pd.read_csv(name)
    data2['tempus']=data2['userID'].map(user_installedapps['userID'].value_counts())
    data2['tempus'].fillna(0,inplace=True)
    data2['tempus']=data2['tempus'].astype(int)
    data2['installedusCounts']=data2['installedusCounts']+data2['tempus']
    
    data2['tempap']=data2['appID'].map(user_installedapps['appID'].value_counts())
    data2['tempap'].fillna(0,inplace=True)
    data2['tempap']=data2['tempap'].astype(int)
    data2['installedapCounts']=data2['installedapCounts']+data2['tempap']
    
    del user_installedapps
    gc.collect()
```

```python
data2.drop(['tempus','tempap'],axis=1,inplace=True)
```
